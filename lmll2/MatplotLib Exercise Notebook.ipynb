## Exercise 4 -  Draw Histogram


```python
data = np.random.randn(1000)
```

```python
plt.hist(data, bins=30, density=True, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Customized Histogram with Density')
plt.show()
```
